```python
from together import Together
import os
import requests
import json
```

```python
os.environ["TOGETHER_API_KEY"] = "e877e4d735cb6ae4e51bb0c4037c12b6a5b5df8eb50ab60e50467a46770bf36f"

def llama32(messages, model_size=11):
  model = f"meta-llama/Llama-3.2-{model_size}B-Vision-Instruct-Turbo"
  url = "https://api.together.xyz/v1/chat/completions"
  payload = {
    "model": model,
    "max_tokens": 4096,
    "temperature": 0.0,
    "stop": ["<|eot_id|>","<|eom_id|>"],
    "messages": messages
  }

  headers = {
    "Accept": "application/json",
    "Content-Type": "application/json",
    "Authorization": "Bearer " + os.environ["TOGETHER_API_KEY"]
  }
  res = json.loads(requests.request("POST", url, headers=headers, data=json.dumps(payload)).content)

  if 'error' in res:
    raise Exception(res['error'])

  return res['choices'][0]['message']['content']
```

```python
messages = [
    {
        "role": "user",
        "content": "How to end climate change?"
    }
]

response = llama32(messages)
print(response)
```

```python
import base64

def encode_image(image_path):
  with open(image_path, "rb") as img:
    return base64.b64encode(img.read()).decode("utf-8")

base64_image = encode_image("/content/global_brazil_drought.webp")
```
